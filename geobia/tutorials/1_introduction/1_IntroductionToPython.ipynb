```python
#Import the GDAL and Matplot libraries.
import gdal
import matplotlib.pyplot as plt
import ipyplot ##This package needs to be installed through anaconda
```

```python
dataset = gdal.Open(r'C:\Users\raechel\Documents\GitHub\raechelportelli.github.io\geobia\tutorials\1_introduction\uas2018nov.tif')
```

```python
print(dataset.RasterCount)
```

```python
band1 = dataset.GetRasterBand(1) # Red channel
band2 = dataset.GetRasterBand(2) # Green channel
band3 = dataset.GetRasterBand(3) # Blue channel
```

```python
b1 = band1.ReadAsArray()
b2 = band2.ReadAsArray()
b3 = band3.ReadAsArray()
```

```python
img = np.dstack((b1, b2, b3))
f = plt.figure()
plt.imshow(img)
plt.savefig('Tiff.png')
plt.show()
```
